```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
from urllib.request import urlopen, Request
```

```python
myurl = "https://stackoverflow.com/questions/415511/how-do-i-get-the-current-time"
```

```python
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36'}
req = Request(myurl, headers=headers)
html = urlopen(req).read()
soupified = BeautifulSoup(html, "html.parser")
```

```python
question_div = soupified.find("div", {"class": "question"})
```

```python
question_div
```

```python
question_text = question_div.get_text(strip=True)
```

```python
question_text
```

```python
answer_divs = soupified.find_all("div", {"class": "answer"})
```

```python
answers = [answer_div.get_text(strip=True) for answer_div in answer_divs]
```

```python
print("Question:")
print(question_text)
print("\nAnswers:")
for idx, answer in enumerate(answers, start=1):
    print(f"Answer {idx}:")
    print(answer)
    print("-" * 40)
```
